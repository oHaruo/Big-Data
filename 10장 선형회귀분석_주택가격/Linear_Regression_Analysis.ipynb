# 데이터 분리하기

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

inplace=False 옵션은 MEDV 속성을 drop했을 때 drop한 속성에 값을 남길 것인지 여부이다

```python
x = data2.drop(['MEDV'], axis = 1,inplace=False)
y = data2['MEDV']
```

random_state 옵션은 '데이터를 어떤 순서로 섞을까'라는 옵션이다. 156이면 156번재 데이터부터 섞인다.

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 156)
print('xtrain 쉐입: ', xtrain.shape)
print('xtest 쉐입: ', xtest.shape)
print('ytrain 쉐입: ', ytrain.shape)
print('ytest 쉐입: ', ytest.shape)
```


*  LinearRegression() 모델 생성
*  lr.fit(xtrain, ytrain) 모델 훈련



```python
lr = LinearRegression()
lr.fit(xtrain, ytrain)
```

```python
y_pred = lr.predict(xtest)
```

평가 지표를 출력해줌 결정계수가 0.6이상이고 RMSE도 3.7정도로 좋은 지표를 보여주고있다

```python
mse = mean_squared_error(ytest, y_pred)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(ytest, y_pred)))
```

```python
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```

회귀식 = 12.58CAT.MEDV + 2.14CHAS + 0.73RM + ************  + -16.32NOX

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index=x.columns)
coef.sort_values(ascending = False)
```

# 산점도, 선형 회귀 그래프

```python
import seaborn as sns
```

```python
fig, axs = plt.subplots(figsize=(16, 16), ncols=3, nrows=5)

x_features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

for i, feature in enumerate(x_features):
      row = int(i/3)
      col = i%3
      sns.regplot(x=feature, y='MEDV', data=data2, ax=axs[row][col])
```
