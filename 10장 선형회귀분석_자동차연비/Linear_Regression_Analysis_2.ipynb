# 데이터 가져오기

```python
import numpy as np
import pandas as pd

data_df = pd.read_csv('auto-mpg.csv', header=0, engine='python')
```

```python
print(' 데이터셋 크기 : ', data_df.shape) #데이터가 어떤 차원인지 알려줌

data_df.head() #데이터를 보여줌
```

분석 안하는 데이터 drop으로 제외

```python
data_df = data_df.drop(['car_name', 'origin', 'horsepower'], axis=1, inplace=False)
```

```python
print(' 데이터세트 크기 : ', data_df.shape) #잘 제외됐는지 확인

data_df.head()
```

```python
data_df.info()
```

# 데이터 분석 모델 구축

```python
from sklearn.linear_model import LinearRegression #LinearRegression 클래스 임포트 - 선형 회귀 모델 객체 만들기
from sklearn.model_selection import train_test_split #학습 데이터를 분리하기 위한 것
from sklearn.metrics import mean_squared_error, r2_score #MSE, R2 계산하기 위한 것
```

```python
# X, Y 분할하기

Y = data_df['mpg']
X = data_df.drop(['mpg'], axis=1, inplace=False)
```

```python
# 훈련용 데이터와 평가용 데이터 분할하기 - train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
# 선형회귀분석 : 모델 생성 - LinearRegression
lr = LinearRegression()
```

```python
# 선형회귀분석 : 모델 훈련
lr.fit(X_train, Y_train)
```

```python
# 선형회귀분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

# 결과 분석및 시각화 (산점도,선형 회귀 그래프)

```python
mse = mean_squared_error(Y_test, Y_predict) #MSE값을 계산
rmse = np.sqrt(mse) #RMSE값을 계산 - MSE값의 제곱근

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict))) #R2값을 계산6
```

```python
print('Y 절편 값: ',  np.round(lr.intercept_, 2))
print('회귀 계수 값: ', np.round(lr.coef_, 2))
```

```python
coef = pd.Series(data=np.round(lr.coef_, 2), index=X.columns) #회귀 계수를 소수점 2번째까지 표현, 인덱스 값은 X의 열
coef.sort_values(ascending=False) #회귀 계수 값을 내림차순으로 표현
```

# 산점도, 선형회귀 그래프로 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(figsize=(16, 16), ncols=3, nrows=2)

x_features = ['model_year', 'acceleration', 'displacement', 'weight', 'cylinders']
plot_color = ['r', 'b', 'y', 'g', 'r']

for i, feature in enumerate(x_features):
      row = int(i/3)
      col = i%3
      sns.regplot(x=feature, y='mpg', data=data_df, ax=axs[row][col], color=plot_color[i])
```
